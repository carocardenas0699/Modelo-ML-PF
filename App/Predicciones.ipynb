```python
prom = pd.read_csv('./prom_mes.csv')
prom
```

```python
# Primero, agregamos una columna "Mes" a la tabla "pred" para extraer el mes de la columna 'Fecha'.
pred['Mes'] = pd.to_datetime(pred['Fecha']).dt.month

# Hacemos un merge entre las dos tablas usando la columna "Mes" en pred y "month" en prom.
pred_combined = pred.merge(prom, left_on='Mes', right_on='month')

pred_combined
```

```python
# Ahora calculamos las nuevas columnas multiplicando cada columna por el valor de "Numero total de viajes".
pred_combined['Numero total de pasajeros'] = pred_combined['Numero total de viajes'] * pred_combined['passenger_trip']
pred_combined['Distancia total recorrida (mi)'] = pred_combined['Numero total de viajes'] * pred_combined['distance_trip']
pred_combined['Ingreso total (USD)'] = pred_combined['Numero total de viajes'] * pred_combined['total_trip']
pred_combined['Duraci贸n total recorrido (min)'] = pred_combined['Numero total de viajes'] * pred_combined['duration_trip']

# Opcional: si ya no necesitas las columnas 'Mes' y 'month', puedes eliminarlas.
pred_combined.drop(columns=['Mes', 'month','passenger_trip', 'distance_trip','total_trip','duration_trip'], inplace=True)

# Visualizamos el resultado
pred_combined
```

```python
pred_combined = pred_combined.round(2)
```

```python
pred_combined['Numero total de pasajeros'] = pred_combined['Numero total de pasajeros'].astype(int)
pred_combined['Numero total de viajes'] = pred_combined['Numero total de viajes'].astype(int)
```

```python
pred_combined
```

```python
pred_combined.to_csv('predicciones.csv')
```
